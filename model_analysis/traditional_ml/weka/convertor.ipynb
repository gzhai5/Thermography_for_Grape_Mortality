```python
import numpy as np
import pandas as pd
import arff
from sklearn.preprocessing import StandardScaler
```

```python
# load data
X_train = np.load(f'../../data-preprocess/data-split/data/features/train/X_train.npy', allow_pickle=True)
y_train = np.load(f'../../data-preprocess/data-split/data/features/train/y_train.npy')
X_test = np.load(f'../../data-preprocess/data-split/data/features/test/X_test.npy', allow_pickle=True)
y_test = np.load(f'../../data-preprocess/data-split/data/features/test/y_test.npy')
print(f'X_train shape: {X_train.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
# Convert list of dicts to DataFrame
df_train = pd.DataFrame([x[0] for x in X_train])
df_train["class"] = y_train

df_test = pd.DataFrame([x[0] for x in X_test])
df_test["class"] = y_test
```

```python
def save_to_arff(df, filename):
    data = {
        'description': '',
        'relation': 'classification_data',
        'attributes': [(col, 'REAL') if col != 'class' else (col, list(map(str, sorted(df['class'].unique())))) for col in df.columns],
        'data': df.values.tolist(),
    }

    with open(filename, 'w') as f:
        arff.dump(data, f)

save_to_arff(df_train, './data/train_data.arff')
save_to_arff(df_test, './data/test_data.arff')
```

```python
# normalize data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(df_train.drop(columns='class'))
X_test_scaled = scaler.transform(df_test.drop(columns='class'))
df_train_scaled = pd.DataFrame(X_train_scaled, columns=df_train.columns[:-1])
df_train_scaled['class'] = df_train['class'].values
df_test_scaled = pd.DataFrame(X_test_scaled, columns=df_test.columns[:-1])
df_test_scaled['class'] = df_test['class'].values

# save normalized data
save_to_arff(df_train_scaled, './data/train_data_normalized.arff')
save_to_arff(df_test_scaled, './data/test_data_normalized.arff')
```
