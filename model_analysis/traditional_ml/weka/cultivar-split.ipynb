```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
# load data
X_train = np.load(f'../../data-preprocess/data-split/data/features/train/X_train.npy', allow_pickle=True)
y_train = np.load(f'../../data-preprocess/data-split/data/features/train/y_train.npy')
X_test = np.load(f'../../data-preprocess/data-split/data/features/test/X_test.npy', allow_pickle=True)
y_test = np.load(f'../../data-preprocess/data-split/data/features/test/y_test.npy')
print(f'X_train shape: {X_train.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
# split the data by cultivars
X_train_ries, y_train_ries, X_train_cf, y_train_cf, X_train_con, y_train_con, X_train_pn, y_train_pn = [], [], [], [], [], [], [], []
for i in range(len(X_train)):
    if X_train[i]['cultivar'] == 1:
        X_train_ries.append(X_train[i])
        y_train_ries.append(y_train[i])
    elif X_train[i]['cultivar'] == 2:
        X_train_cf.append(X_train[i])
        y_train_cf.append(y_train[i])
    elif X_train[i]['cultivar'] == 3:
        X_train_con.append(X_train[i])
        y_train_con.append(y_train[i])
    elif X_train[i]['cultivar'] == 4:
        X_train_pn.append(X_train[i])
        y_train_pn.append(y_train[i])
X_test_ries, y_test_ries, X_test_cf, y_test_cf, X_test_con, y_test_con, X_test_pn, y_test_pn = [], [], [], [], [], [], [], []
for i in range(len(X_test)):
    if X_test[i]['cultivar'] == 1:
        X_test_ries.append(X_test[i])
        y_test_ries.append(y_test[i])
    elif X_test[i]['cultivar'] == 2:
        X_test_cf.append(X_test[i])
        y_test_cf.append(y_test[i])
    elif X_test[i]['cultivar'] == 3:
        X_test_con.append(X_test[i])
        y_test_con.append(y_test[i])
    elif X_test[i]['cultivar'] == 4:
        X_test_pn.append(X_test[i])
        y_test_pn.append(y_test[i])

X_train_ries = np.array(X_train_ries)
y_train_ries = np.array(y_train_ries)
X_train_cf = np.array(X_train_cf)
y_train_cf = np.array(y_train_cf)
X_train_con = np.array(X_train_con)
y_train_con = np.array(y_train_con)
X_train_pn = np.array(X_train_pn)
y_train_pn = np.array(y_train_pn)
X_test_ries = np.array(X_test_ries)
y_test_ries = np.array(y_test_ries)
X_test_cf = np.array(X_test_cf)
y_test_cf = np.array(y_test_cf)
X_test_con = np.array(X_test_con)
y_test_con = np.array(y_test_con)
X_test_pn = np.array(X_test_pn)
y_test_pn = np.array(y_test_pn)

print(f'X_train_ries shape: {X_train_ries.shape}')
print(f'y_train_ries shape: {y_train_ries.shape}')
print(f'X_train_cf shape: {X_train_cf.shape}')
print(f'y_train_cf shape: {y_train_cf.shape}')
print(f'X_train_con shape: {X_train_con.shape}')
print(f'y_train_con shape: {y_train_con.shape}')
print(f'X_train_pn shape: {X_train_pn.shape}')
print(f'y_train_pn shape: {y_train_pn.shape}')
print(f'X_test_ries shape: {X_test_ries.shape}')
print(f'y_test_ries shape: {y_test_ries.shape}')
print(f'X_test_cf shape: {X_test_cf.shape}')
print(f'y_test_cf shape: {y_test_cf.shape}')
print(f'X_test_con shape: {X_test_con.shape}')
print(f'y_test_con shape: {y_test_con.shape}')
print(f'X_test_pn shape: {X_test_pn.shape}')
print(f'y_test_pn shape: {y_test_pn.shape}')
```

```python
# store the data
np.save(f'./data/ries/X_train_ries.npy', X_train_ries)
np.save(f'./data/ries/y_train_ries.npy', y_train_ries)
np.save(f'./data/cf/X_train_cf.npy', X_train_cf)
np.save(f'./data/cf/y_train_cf.npy', y_train_cf)
np.save(f'./data/conc/X_train_con.npy', X_train_con)
np.save(f'./data/conc/y_train_con.npy', y_train_con)
np.save(f'./data/pn/X_train_pn.npy', X_train_pn)
np.save(f'./data/pn/y_train_pn.npy', y_train_pn)
np.save(f'./data/ries/X_test_ries.npy', X_test_ries)
np.save(f'./data/ries/y_test_ries.npy', y_test_ries)
np.save(f'./data/cf/X_test_cf.npy', X_test_cf)
np.save(f'./data/cf/y_test_cf.npy', y_test_cf)
np.save(f'./data/conc/X_test_con.npy', X_test_con)
np.save(f'./data/conc/y_test_con.npy', y_test_con)
np.save(f'./data/pn/X_test_pn.npy', X_test_pn)
np.save(f'./data/pn/y_test_pn.npy', y_test_pn)
```
