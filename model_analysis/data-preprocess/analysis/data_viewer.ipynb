```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_data_path = "./CF6.0_B01_N01.npy"
data = np.load(input_data_path)
print(data.shape)

critical_frames = [0, 149, 299, 449, 599]
```

```python
def visualize_frames(data):
    for frame in critical_frames:
        plt.imshow(data[frame], cmap='gray')
        plt.axis('off')
        plt.subplots_adjust(0, 0, 1, 1)
        # plt.show()
        plt.savefig(f"./frames/frame_{frame}.png", bbox_inches='tight', pad_inches=0, dpi=500)
```

```python
def normalize_curve_mean_method1(input_data):
    mean = np.mean(input_data[:30])
    normalized_data = np.maximum(input_data - mean, 0)
    return normalized_data

normalized_data = normalize_curve_mean_method1(data)
```
