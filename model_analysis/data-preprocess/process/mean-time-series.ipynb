```python
import numpy as np
```

```python
total_video_data = np.load('../data/roi-video/total_data.npy', allow_pickle=True).item()
print('total video data len:', len(total_video_data), 'shape:', list(total_video_data.values())[0].shape)
```

```python
total_data = {}
for k, v in total_video_data.items():
    total_data[k] = np.mean(v, axis=(1, 2))
print('total data len:', len(total_data), 'shape:', list(total_data.values())[0].shape)
```

```python
np.save('../data/mean-time-series/total_data.npy', total_data)
```
