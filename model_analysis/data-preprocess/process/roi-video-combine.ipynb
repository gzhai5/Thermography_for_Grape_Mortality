```python
import numpy as np
```

```python
exp4_data = np.load(f'../data/roi-video/exp_4_total_data.npy', allow_pickle=True).item()
exp5_data = np.load(f'../data/roi-video/exp_5_total_data.npy', allow_pickle=True).item()
exp4_label = np.load(f'../data/roi-video/exp_4_total_label.npy', allow_pickle=True).item()
exp5_label = np.load(f'../data/roi-video/exp_5_total_label.npy', allow_pickle=True).item()
exp4_cultivar = np.load(f'../data/roi-video/exp_4_total_cultivar.npy', allow_pickle=True).item()
exp5_cultivar = np.load(f'../data/roi-video/exp_5_total_cultivar.npy', allow_pickle=True).item()

print('exp#4 data len:', len(exp4_data), 'shape:', list(exp4_data.values())[0].shape)
print('exp#5 data len:', len(exp5_data), 'shape:', list(exp5_data.values())[0].shape)
print('---------------------------------')
print('exp#4 label len:', len(exp4_label))
print('exp#5 label len:', len(exp5_label))
print('---------------------------------')
print('exp#4 label len:', len(exp4_label))
print('exp#5 data len:', len(exp5_cultivar))
```

```python
# combine data
total_data = exp4_data | exp5_data
total_label = exp4_label | exp5_label
total_cultivar = exp4_cultivar | exp5_cultivar

print('total data len:', len(total_data), 'shape:', list(total_data.values())[0].shape)
print('total label len:', len(total_label))
print('total cultivar len:', len(total_cultivar))
```

```python
# save data
np.save('../data/roi-video/total_data.npy', total_data)
np.save('../data/roi-video/total_label.npy', total_label)
np.save('../data/roi-video/total_cultivar.npy', total_cultivar)
```
