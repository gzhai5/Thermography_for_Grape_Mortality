```python
from tensorflow.keras.utils import plot_model
from lstmfcn import generate_lstmfcn
from lstm import generate_lstm
```

```python
m1 = generate_lstm(feature_input_accept=False, print_summary=True)
m2 = generate_lstm(feature_input_accept=True, print_summary=True)
m3 = generate_lstmfcn(feature_input_accept=False, print_summary=True)
m4 = generate_lstmfcn(feature_input_accept=True, print_summary=True)

# Save SVG/PNG with shapes
plot_model(m1, to_file="lstm.png", show_shapes=True, expand_nested=True, dpi=200)
plot_model(m2, to_file="lstm_plus_features.png", show_shapes=True, expand_nested=True, dpi=200)
plot_model(m3, to_file="lstm_fcn.png", show_shapes=True, expand_nested=True, dpi=200)
plot_model(m4, to_file="lstm_fcn_plus_features.png", show_shapes=True, expand_nested=True, dpi=200)
```
