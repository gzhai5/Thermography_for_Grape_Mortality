```python
import os
import numpy as np
import pandas as pd
```

```python
def extract_frames(data: np.ndarray, y: int, x: int, size: int) -> np.ndarray:
    half = size // 2
    if size % 2 == 0:
        y_start, y_end = y - half, y + half
        x_start, x_end = x - half, x + half
    else:
        y_start, y_end = y - half, y + half + 1
        x_start, x_end = x - half, x + half + 1

    extracted_frames = []
    for frame in data:
        extracted_frame = frame[y_start:y_end, x_start:x_end]
        extracted_frames.append(extracted_frame)

    return np.array(extracted_frames)
```

```python
size = 21
```

```python
df_new = pd.read_csv('./csv/stat_all_data.csv')

# traverse all rows, for each row, get the npy file, extract the roi data, and store the data
for index, row in df_new.iterrows():
    filename = row['filename']
    round_number = row['dev_stage']
    x_pt = row['x_coordinate']
    y_pt = row['y_coordinate']

    if round_number == 5 or round_number == 6:

        raw_data = np.load(f'D:/Bud Thermography Data/round_{round_number}/{filename}')
        roi_data = extract_frames(raw_data, y_pt, x_pt, size)

        print('processing', filename)
        print('round_number', round_number)
        print('shape of raw_data:', raw_data.shape)
        print('shape of roi_data:', roi_data.shape)

        # Ensure the directory exists
        output_dir = f'./size-{size}/round_{round_number}/'
        os.makedirs(output_dir, exist_ok=True)

        # Save the ROI data
        np.save(f'{output_dir}[ROI]{filename}', roi_data)
        print(f'saved roi data to ./size-{size}/round_{round_number}/[ROI]{filename}')
        print('-------------------------------------------------')
```
