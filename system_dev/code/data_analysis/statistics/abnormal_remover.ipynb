```python
from abnormals import abnormal_samples_set
import pandas as pd
```

```python
# read the data
df = pd.read_csv("/Users/mud/Desktop/Thermography_for_Grape_Mortality/code/data_analysis/statistics/csv/stat_all_data.csv")
print("Number of rows in original data: ", df.shape[0])

# filter the data, remove all rows whose filename col value is in the abnormal_samples_set
# value in abnormal_samples_set is a list of filenames but looks like this: r"D:\Bud Thermography Data\round_6\PN6.0_B27_N05.npy"
# while the filename col value looks like this: "PN6.0_B27_N05.npy"
# so we need to extract the filename from the path in abnormal_samples_set
abnormal_samples_set = [filename.split("\\")[-1] for filename in abnormal_samples_set]
print("Number of abnormal samples: ", len(abnormal_samples_set))

cantremoved = []
for filename in abnormal_samples_set:
    if "RIES" in filename:
        filename = filename.replace("RIES", "Ries")
    if "CON" in filename:
        filename = filename.replace("CON", "Con")
    if "Ries3.0" in filename:
        filename = filename.replace("Ries3.0", "RIES3.0")
    if 'Con5.0_B19_N09.npy' in filename:
        filename = filename.replace("Con5.0_B19_N09.npy", "con5.0_B19_N09.npy")
    if 'Con5.0_B24_N02.npy' in filename:
        filename = filename.replace("Con5.0_B24_N02.npy", "con5.0_B24_N02.npy")

    if filename in df["filename"].values:
        df = df[df["filename"] != filename]
    else:
        cantremoved.append(filename)


# check the whole number removed
print("Number of abnormal samples not removed: ", len(cantremoved))
print("Number of rows now: ", df.shape[0])
```

```python
df.to_csv("/Users/mud/Desktop/Thermography_for_Grape_Mortality/code/data_analysis/statistics/csv/stat_all_data_noabnormal.csv", index=False)
```
