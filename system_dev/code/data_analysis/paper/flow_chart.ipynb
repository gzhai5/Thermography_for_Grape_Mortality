```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
data = np.load('/Volumes/My Book Duo/Bud Thermography Data/round_6/CF6.0_B10_N06.npy')
print(data.shape)
frames, height, width = data.shape
```

```python
frame_idxs = [0, 100, 300, 400, 599]

for frame_idx in frame_idxs:
    frame = data[frame_idx]

    # Plot the frame
    plt.figure(figsize=(6, 6))
    plt.imshow(frame, cmap='gray')
    plt.axis('off')  # Remove axes
    plt.gca().set_xticks([])  # Remove x-ticks
    plt.gca().set_yticks([])  # Remove y-ticks
    plt.tight_layout()

    # Save the plot
    plt.savefig(f'orginal_data_frame_{frame_idx}.svg', bbox_inches='tight', pad_inches=0)

    plt.show()
```

```python
roi_center = [323, 233]
roi_size = [21, 21]

for frame_idx in frame_idxs:
    frame = data[frame_idx]

    # Extract the region of interest
    roi = frame[roi_center[1] - roi_size[1] // 2:roi_center[1] + roi_size[1] // 2,
                roi_center[0] - roi_size[0] // 2:roi_center[0] + roi_size[0] // 2]

    # Plot the region of interest inside the frame with a red rectangle
    plt.figure(figsize=(6, 6))
    plt.imshow(frame, cmap='gray')
    plt.gca().add_patch(plt.Rectangle((roi_center[0] - roi_size[0] // 2, roi_center[1] - roi_size[1] // 2),
                                    roi_size[0], roi_size[1], linewidth=1, edgecolor='r', facecolor='none'))
    plt.axis('off')  # Remove axes
    plt.gca().set_xticks([])  # Remove x-ticks
    plt.gca().set_yticks([])  # Remove y-ticks
    plt.tight_layout()

    # Save the plot
    plt.savefig(f'roi_frame_{frame_idx}.svg', bbox_inches='tight', pad_inches=0)

    plt.show()
```
